```python
import pandas as pd
import requests
import json
import numpy as np
import pickle as pkl
import os
```

```python
books_in=pd.read_csv("C:\\Users\\anlan\\Documents\\School Stuff\\2nd Year\\3240\\books_google.csv")
books_in=books_in.dropna(subset=['ISBN'])
books_in=books_in.astype(str)
```

```python
h={'Authorization': '43917_10d89d554a60d00b50deb146d2acc0a5'}
for index in books_in.index:
    status_code=1
    isbn=books_in.at[index, "ISBN"]
    if (os.path.exists(f"C:\\Users\\anlan\\Documents\\School Stuff\\2nd Year\\3240\\Response Pkls\\isbndb\\{isbn.replace('-','')}.pkl")):
        with open(f"C:\\Users\\anlan\\Documents\\School Stuff\\2nd Year\\3240\\Response Pkls\\isbndb\\{isbn.replace('-','')}.pkl", 'rb') as f_in:
            try:
                response_json=pkl.load(f_in)
            except EOFError:
                print("404 for this isbn: "+isbn)
                continue
    else:
        continue
#         response = requests.get(f"https://api2.isbndb.com/book/{isbn.replace('-','')}", headers=h)
#         if (response.status_code!=200):
#             print("got status code "+response.status_code+" for isbn: "+isbn)
#             print(index)
#             continue
#         with open(f"C:\\Users\\anlan\\Documents\\School Stuff\\2nd Year\\3240\\Response Pkls\\isbndb\\{isbn.replace('-','')}.pkl", 'wb') as f:
#             try:
#                 response_json=response.json()
#                 pkl.dump(response_json, f)
#             except:
#                 continue
    try:
        books_in.at[index, 'Title']=response_json['book']['title_long']
    except KeyError:
        try:
            books_in.at[index, 'Title']=response_json['book']['title']
        except KeyError:
            print("no title: "+isbn)
            continue
    books_in['Author']=books_in['Author'].astype(object)
    try:
        books_in.at[index, 'Author']=response_json['book']['authors']
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'Publisher']=response_json['book']['publisher']
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'Publish Year']=response_json['book']['date_published']
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'Description']=response_json['book']['synopsys']
    except KeyError:
        pass

    books_in['ISBN10']=books_in['ISBN10'].astype(str)
    books_in['ISBN13']=books_in['ISBN13'].astype(str)
    try:
        books_in.at[index, 'ISBN13']=str(response_json['book']['isbn13'])
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'ISBN10']=str(response_json['book']['isbn'])
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'Page Count']=response_json['book']['pages']
    except KeyError:
        pass
    
    books_in['Image Links']=books_in['Image Links'].astype(object)
    
    try:
        books_in.at[index, 'Image Links']=response_json['book']['image']
    except KeyError:
        pass
    
    try:
        books_in.at[index, 'Overview']=response_json['book']['overview']
    except KeyError:
        pass
    
```

```python
books_in.to_json(f"C:\\Users\\anlan\\Documents\\School Stuff\\2nd Year\\3240\\books_with_isbndb.json", orient='records')
```
